```python
import json
import pandas as pd

from pathlib import Path
```

```python
results_path = Path('results/processing_results.json')

# Load the JSON data
data = json.load(results_path.open('r'))

# Navigate to the specific dictionary
# data[0] gets the first dictionary in the outer list
# ["image_analysis"] accesses the "image_analysis" list
# [1] gets the second element (dictionary) within the "image_analysis" list
try:
    target_dictionary = data[0]["image_analysis"][1]
    
    # Extract the keys (headers)
    headers = list(target_dictionary.keys())
    
    print("Extracted Headers:")
    for header in headers:
        print(f"- {header}")

except (IndexError, KeyError) as e:
    print(f"Error accessing data: {e}")
    print("Please ensure the JSON structure matches the expected path (data[0]['image_analysis'][1]).")
```

```python
desired_fields = [
    "ROI-label",
    "Analysis-method",
    ("amplitude", "Mean"),
    ("mu", "Mean"),
    ("offset", "Mean"),
    ("error", "Mean")
]

# Prepare a list to hold rows of data
records = []

# Iterate through each top-level dictionary in the JSON data (each "manifest_entry")
for entry in data:
    # Extract the image_timestamp once per entry
    image_timestamp = entry.get("image_timestamp")

    temp_sensors = entry.get("temperature_readings", {}).keys()
    temp_dict = {}
    for sensor in temp_sensors:
        temp_dict[f'temperature_{sensor.lower().replace(' ', '_')}'] = entry.get("temperature_readings", {}).get(sensor, None)

    # Iterate through each item in the "image_analysis" list
    for analysis_item in entry.get("image_analysis", []):
        # We are interested in elements that have "ROI-label" as a key
        if "ROI-label" in analysis_item:
            row_data = {}
            # Add the image_timestamp to the current row_data
            row_data["timestamp"] = image_timestamp

            for field in desired_fields:
                # Skip if the field is image_timestamp as it's already added
                if field == "image_timestamp":
                    continue

                if isinstance(field, tuple):
                    # It's a nested field like ("amplitude", "Mean")
                    outer_key, inner_key = field
                    value = analysis_item.get(outer_key, {}).get(inner_key)
                    row_data[outer_key] = value
                else:
                    # It's a direct field like "ROI-label" or "Analysis-method"
                    value = analysis_item.get(field)
                    row_data[field] = value
            row_data.update(temp_dict)
            records.append(row_data)

# Create the DataFrame
df = pd.DataFrame(records)

print("DataFrame created successfully:")
print(df)

# To verify headers (optional, as they are constructed from desired_fields)
print("\nDataFrame Headers:")
print(df.columns.tolist())
```
